```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('netflix_titles.csv')
```

```python
df.head(10)
```

```python
df.info
```

```python
df.columns
```

```python
#Rename column headers to be clean and uniform
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
total_duplicates = df.duplicated().sum()
print(f"Total number of exact duplicate rows: {total_duplicates}")
```

```python
# Before 
null= df.isnull().sum()
print(null)
```

```python
# handling null vALUES
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

df.dropna(subset=['rating', 'duration', 'date_added'], inplace=True)
```

```python
#after
null= df.isnull().sum()
print(null)
```

```python
df.dtypes
```

```python
df.head()
```

```python
# change date format
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y').dt.strftime('%d-%m-%Y')
```

```python
df.head()
```

```python
df.to_csv('cleaned_netflix_titles.csv', index=False)
```
